```python
df = pd.read_csv('data.csv')
```

```python
import requests
from bs4 import BeautifulSoup

clg_names = []

def url_generator(college_name):
    url = 'https://www.google.com/search?q='
    college_name = college_name.replace(' ', '+')
    college_name = college_name.replace(',', '%2C')
    college_name = college_name.replace('&', '%26')
    url = url + college_name + ' address'
        
    return url

def find_address(clg_name):
    url = url_generator(clg_name)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    try:
        html_text = soup.find_all(class_='BNeawe iBp4i AP7Wnd')
        print(html_text[0].text)
        clg_names.append(html_text[0].text)
    except Exception as e:
        clg_names.append(0)

for i in df.index:
    print(i)
    find_address(df.iloc[i]['COLLEGE NAME'])
```

```python
clg_names
```

```python
for i in df.index:
    df.at[i,'ADRESS'] = clg_names[i]
```
